## Conditions multiples

Rappelez-vous les tapes :

* Obtenir les conditions
* Encadrer chaque condition par des parenthses
* Utilisez l'oprateur | ou &, selon que vous voulez un rsultat de type 
    * OR | (l'une ou l'autre des conditions est vraie)
    * AND & (les deux conditions doivent tre vraies)
* Vous pouvez galement utiliser l'oprateur ~ comme une opration NOT.

On veut avoir la liste des hommes avec bill plus grand que 30

```python
(df["total_bill"] > 30) & (df['sex']=='Male')
```

```python
df[(df["total_bill"] > 30) & (df['sex']=='Male')].head()
```

```python
len(df[(df["total_bill"] > 30) & (df['sex']=='Male')])
```

On veut la liste des tous sauf homme avec bill plus grand que 30

```python
(df['total_bill'] > 30) & (df['sex']!='Male')
```

```python
df[(df['total_bill'] > 30) & (df['sex']!='Male')]
```

```python
len(df[(df['total_bill'] > 30) & (df['sex']!='Male')])
```

On veut la liste des tous sauf homme avec bill plus grand que 30

```python
(df['total_bill'] > 30) & ~(df['sex']=='Male')
```

```python
df[(df['total_bill'] > 30) & ~(df['sex']=='Male')]
```

```python
len(df[(df['total_bill'] > 30) & ~(df['sex']=='Male')])
```

On veut avoir tous les lignes de samedi et dimanche seulement

```python
(df['day'] =='Sun') | (df['day']=='Sat')
```

```python
df[(df['day'] =='Sun') | (df['day']=='Sat')]
```

```python
len(df[(df['day'] =='Sun') | (df['day']=='Sat')])
```

## Oprateur conditionnel isin()

Nous pouvons utiliser l'oprateur .isin() pour filtrer par une liste d'options.

```python
#On veut avoir tous les lignes de samedi et dimanche seulement
options = ['Sat','Sun']
df['day'].isin(options)
```

```python
df[df['day'].isin(options)]
```

```python
len(df[df['day'].isin(options)])
```

```python
df['day'].isin(['Sat','Sun']) #Methode 2
```

```python
df[df['day'].isin(['Sat','Sun'])]
```

```python
len(df[df['day'].isin(['Sat','Sun'])])
```
