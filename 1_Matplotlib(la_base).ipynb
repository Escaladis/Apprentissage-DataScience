## Les bases Matplotlib

### Introduction

Matplotlib est la bibliothèque « grand-père » de la visualisation de données avec Python. Il a été créé par John Hunter. Il l'a créé pour essayer de répliquer les capacités de tracé de MatLab (un autre langage de programmation) en Python. Donc, si vous êtes familier avec matlab, matplotlib vous semblera naturel.


Il s'agit d'une excellente bibliothèque graphique 2D et 3D pour générer des figures scientifiques.


Voici quelques-uns des principaux avantages de Matplotlib :

* Généralement facile pour démarrer avec des tracés simples
* Prise en charge des étiquettes et des textes personnalisés
* Excellent contrôle de chaque élément d'une figure
* Sortie de haute qualité dans de nombreux formats
* Très personnalisable en général

Matplotlib vous permet de créer des figures reproductibles par programmation. Apprenons comment l'utiliser ! Avant de poursuivre cette leçon, je vous encourage simplement à explorer la page Web officielle de Matplotlib : http://matplotlib.org/

## Exemple de base

Prenons un exemple très simple en utilisant deux tableaux NumPy :

### Tracé de base d'un tableau

Voyons un exemple très simple en utilisant deux tableaux NumPy. Vous pouvez également utiliser des listes, mais il est fort probable que vous transmettiez des tableaux NumPy ou des colonnes Pandas (qui se comportent essentiellement comme des tableaux).


**Les données que nous voulons tracer :**

```python
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
x
```

```python
y
```

## Utilisation de Matplotlib avec les appels de la fonction plt.plot()

### Commandes Matplotlib de base


```python
plt.plot(x, y) #x est la barre horizontale, y est la barre verticale
plt.xlabel("Titre de l'axe X ici") #titre barre horizontale
plt.ylabel("Titre de l'axe Y ici") #titre barre verticale
plt.title('Titre général ici') #le titre
plt.show() # Requis pour les utilisateurs non-jupyter, mais supprime également l'information Out[].
```

### Modification d'autres paramètres de la figure

```python
plt.plot(x, y) 
plt.xlabel("Titre de l'axe X ici")
plt.ylabel("Titre de l'axe Y ici")
plt.title('Titre général ici')
plt.xlim(0,5) # Limite inférieure, limite supérieure
plt.ylim(0,10) # Limite inférieure, limite supérieure
plt.show() # Requis pour les utilisateurs non-jupyter, mais supprime également l'information Out[].
```
