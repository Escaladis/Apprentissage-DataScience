### Trac de base d'un tableau

Voyons un exemple trs simple en utilisant deux tableaux NumPy. Vous pouvez galement utiliser des listes, mais il est fort probable que vous transmettiez des tableaux NumPy ou des colonnes Pandas (qui se comportent essentiellement comme des tableaux).

## Utilisation de Matplotlib avec les appels de la fonction plt.plot()

### Commandes Matplotlib de base


```python
plt.plot(x, y) #x est la barre horizontale, y est la barre verticale
plt.xlabel("Titre de l'axe X ici") #titre barre horizontale
plt.ylabel("Titre de l'axe Y ici") #titre barre verticale
plt.title('Titre gnral ici') #le titre
plt.show() # Requis pour les utilisateurs non-jupyter, mais supprime galement l'information Out[].
```

### Modification d'autres paramtres de la figure

```python
plt.plot(x, y) 
plt.xlabel("Titre de l'axe X ici")
plt.ylabel("Titre de l'axe Y ici")
plt.title('Titre gnral ici')
plt.xlim(0,5) # Limite infrieure, limite suprieure
plt.ylim(0,10) # Limite infrieure, limite suprieure
plt.show() # Requis pour les utilisateurs non-jupyter, mais supprime galement l'information Out[].
```
