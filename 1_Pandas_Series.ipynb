# Series

Une série est très similaire à un tableau NumPy (en fait, il est construit au-dessus de l'objet array NumPy). Ce qui différencie le tableau NumPy d'une série, c'est qu'une série peut comporter des étiquettes d'axe, ce qui signifie qu'elle peut être indexée par une étiquette, au lieu d'un seul emplacement numérique. Il n'a pas non plus besoin de contenir des données numériques, il peut contenir n'importe quel objet Python arbitraire.

### Importations

```python
import numpy as np
import pandas as pd
```

## Création d'une série à partir d'objets Python

### Listes d'index et de données


Nous pouvons créer une série à partir de listes Python (également à partir de tableaux NumPy)

```python
# help(pd.Series)
```

```python
pd.Series #(shift + tab pour plus d'informations)
```

```python
myindex = ['USA','Canada','Mexico'] #liste python
```

```python
my_data = [1776, 1867, 1821] #liste python
```

```python
my_ser = pd.Series(data = my_data) #data : ajouter les données en forme de serie
my_ser
```

```python
type(my_ser) #le type de données est Series
```

```python
my_ser = pd.Series(data = my_data, index = myindex) #index: ajouter un index
my_ser
```

```python
#Il n'est pas necessaire d'indiquer data, index.
#pandas comprend que la premiere sera les données et la deuxieme l'index
my_ser = pd.Series(my_data, myindex)
my_ser
```

```python
#on cherche la ligne avec l'index 0
my_ser[0] 
```

```python
#on cherche la ligne avec l'index "USA"
my_ser["USA"] 
```

###  Création d'une série à partir d'un dictionnaire

```python
ages = {"Sam":5, "Frank":10, "Spike":7,} #creation d'un dictionnaire en python
```

```python
pd.Series(ages) #pandas transforme automatiquement un dictionnaire dans une serie
```

## Idées clés d'une Série

### Index nominatif

```python
# Données de ventes (inventés) pour les 1er et 2e trimestres pour une entreprise mondiale
q1 = {'Japan': 80, 'China': 450, 'India': 200, 'USA': 250}
q2 = {'Brazil': 100,'China': 500, 'India': 210,'USA': 260}
```

```python
# Convertir en série Pandas
sales_q1 = pd.Series(q1)
sales_q2 = pd.Series(q2)
```

```python
#le nom de pays est egalement l'index
sales_q1 
```

```python
#le nom de pays est egalement l'index
sales_q2 
```

```python
# Appels de valeurs basés sur l'index nominatif
sales_q1["Japan"]
```

```python
# Les informations de localisation basées sur des entiers sont également conservées !
sales_q1[0]
```

**Soyez prudent avec les erreurs potentielles !**

```python
# Faux nom
# sales_Q1['France']
```

```python
# Espace supplémentaire accidentel
# sales_Q1['USA ']
```

```python
# Erreur de majuscule
# sales_Q1['usa']
```

### Opérations

```python
# Saisir uniquement les clés d'index
sales_q1.keys()
```

```python
# Peut effectuer des opérations diffusées sur l'ensemble de la série
sales_q1 * 2
```

```python
[1, 2] * 2 #python ne peux pas multiplier les listes
```

```python
np.array([1, 2]) * 2 #mais numpy peut et pandas aussi car pandas a été crée a partir de numpy
```
